```python
df['Prediction'].value_counts().plot(kind = 'bar', figsize=(12,12), \
                                         color = 'red')
plt.show
```

```python
df['Prediction'].value_counts()[:2].plot(kind='pie', figsize=(12,12), \
                                         colors=['red', 'blue'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
ax.pie(df['Prediction'].value_counts()[:2], labels=df['Prediction'].value_counts()[:2].index, colors=['red', 'blue'])
ax.set_title('My Pie Chart')
plt.show()
```

```python
# Create pie chart
fig, ax = plt.subplots(figsize=(12, 12))
ax.pie(df['Prediction'].value_counts()[:2], colors=['red', 'blue'])

# Add a legend
labels = ['not spam', 'spam']
handles = [plt.Rectangle((0,0),1,1, color=color) for color in ['red', 'blue']]
ax.legend(handles, labels, loc='best')

# Set the title
ax.set_title('My Pie Chart')

plt.show()
```

```python

sns.catplot(x = 'Prediction', data = df, kind = 'count')
plt.title('Count based on spam classifications')
plt.show()
```

```python
spam_count = df['Prediction'].value_counts()


label = ['Not_spam', 'Spam']
explode = [0, 0.1]

palette_color = sns.color_palette('dark')

plt.pie(spam_count, labels = label, colors=palette_color, explode = explode )
plt.title('Proportion of Spams and Non-Spam Mails')
plt.show()
```

```python
cols=['the', 'ect', 'for', 'and', 'you', 'hou', 'com', 'our', 'his', 'her','Prediction']

sns.heatmap(df[cols].corr(),annot=True);
```
